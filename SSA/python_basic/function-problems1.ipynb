### 문제1

>2016년 1월 1일은 금요일입니다. 2016년 a월 b일은 무슨 요일일까요? 두 수 a,b를 입력받아 2016년 a월 b일이 무슨 요일인지 리턴하는 함수, solution을 완성하세요. 요일의 이름은 일요일부터 토요일까지 각각 SUN,MON,TUE,WEB,THU,FRI,SAT입니다. 

>예를 들어, a=5,b=24라면 5월 24일은 화요일이므로 문자열 "TUE"를 반환하세요. 

```python
#여기에 코드를 작성하세요
import datetime

def getWeekDay(a,b):
    weekday=["MON","TUE","WEB","THU","FRI","SAT","SUN"]
    the_time=datetime.datetime(2016,a,b)
    day=the_time.weekday()
    return weekday[day]

getWeekDay(1,1),getWeekDay(5,24)
```

### 문제2

>어떤 문장의 각 알파벳을 일정한 거리만큼 밀어서 다른 알파벳으로 바꾸는 암호화 방식을 시저 암호라고 합니다. 
예를 들어 AB는 1만큼 밀면 BC가 되고, 3만큼 밀면 DE가 됩니다. z는 1만큼 밀면 a가 됩니다. 
문자열 s와 거리 n을 입력받아 s를 n만큼 민 암호문을 만드는 함수, solution을 완성해 보세요.


>제한 조건: 공백은 아무리 밀어도 공백. s는 알파벳 소문자, 대문자, 공백으로만 이루어져 있다. 
s의 길이는 8000이하이다. n은 1이상 25이하인 자연수이다.

```python
# 여기에 코드를 작성하세요
class Code:
    def __init__(self):
        self.alph="abcdefghijklmnopqrstuvwxyz"*2
    
    def encoding(self,s,n):
        encoded=""
        upper_memor=self.upper_idx(s)
        s=s.lower()
        for index,letter in enumerate(s):
            if not letter==" ":
                idx=self.alph.find(letter)
                if index in upper_memor:
                    encoded+=self.alph[idx+n].upper()
                else:
                    encoded+=self.alph[idx+n]
            else:
                encoded+=" "
        return encoded
    
    def decoding(self,s,key):
        decoded=""
        upper_memor=self.upper_idx(s)
        s=s.lower()
        for index,letter in enumerate(s):
            if not letter==" ":
                idx=26+self.alph.find(letter)
                if index in upper_memor:
                    decoded+=self.alph[idx-key].upper()
                else:
                    decoded+=self.alph[idx-key]
            else:
                decoded+=" "
        return decoded
    def upper_idx(self,s):
        upper_memor=[]
        for idx,letter in enumerate(s):
            if (letter!=" ") and ord(letter)<91:
                upper_memor.append(idx)
        return upper_memor
```

```python
s="Hi My name is JiMyeong G C Son"
sizor=Code()
key=6
print(s)
encoded=sizor.encoding(s,key)
print(f"encoded: {encoded}")

decoded=sizor.decoding(encoded,key)
print(f"decoded: {decoded}")

misdecoded=sizor.decoding(encoded,8)
print("using wrong key")
print(misdecoded)
```

>제한 사항: 문자열 전체의 짝/홀수 인덱스가 아니라, 단어(공백을 기준)별로 짝/홀수 인덱스를 판단해야 합니다.


```python
# 여기에 코드를 작성하세요
example ="apple banna   why    Who are  you"

def changer(s):
    flag=0
    result=""
    s=s.lower()
    for word in s:
        if word==" ":
            result+=" "
            flag=0
        else:
            flag+=1
            if flag==2:
                result+=word.upper()
            else:
                result+=word
    return result

result=changer(example)
print(result)
```

### 문제5

>당신은 리스트의 요소중에서 처음으로 연속되지 않은 요소를 찾아야 합니다.예를 들어, [1, 2, 3, 4, 6, 7, 8] 이라는 리스트에서 1, 2, 3, 4는 모두 연속적이지만 6은 처음으로 연속되지 않은 요소입니다. 만약에 모든 리스트의 요소가 연속적이라면 null을 반환합니다. 배열에는 최소한 2개의 요소가 있어야 하고 모든 요소의 값은 숫자이며 중복은 없고 오름차순으로 정렬됩니다. 숫자는 음수 또는 양수 모두 가능합니다.

```python
#여기에 코드를 작성하세요
#1
def continuity_check(check_list):
    remember=check_list.pop(0)
    for num in check_list:
        if remember+1!=num:
            return 0
        else:
            remember+=1
    return "null"

#2 
import numpy as np
def continuity_check_np(check_list):
    a=np.array(check_list)
    b=a-a.min()
    
    for idx,value in enumerate(b):
        if abs(idx-value)>1e-10:
            return 0
    return "null"
            
#3
import copy
def continuity_check_np2(check_list):
    check_list2=copy.deepcopy(check_list)
    firstValue=check_list2.pop(0)
    lastValue=check_list.pop()
    
    a=np.array(check_list)
    b=np.array(check_list2)
    
    c=b-a-1
    if not c.nonzero()[0].tolist():
        if firstValue==a[0] and lastValue==b[-1]:
            return "null"
    else:
        return 0
```

```python
chklist1=[-2,-1,0,1,2,3,4,5,6,7,8]
chklist2=[-2,-1,0,1,2,1,4,2,6,7,8]

result0=continuity_check(chklist1)
print(result0)
result1=continuity_check(chklist2)
print(result1)

result1=continuity_check_np(chklist1)
print(result1)
result2=continuity_check_np(chklist2)
print(result2)

result3=continuity_check_np2(chklist1)
print(result3)
result4=continuity_check_np2(chklist2)
print(result4)
```
