# I. Disjoint Set

```python
# DisJoint_Set by list1


def make_set(p, x):
    p[x] = x

def find_set(p, x):
    if p[x] == x:
        return x
    else:
        val = find_set(p, p[x])
        p[x] = val
        return val

def union_set(p, x, y):
    p[find_set(p, y)] = find_set(p, x)

p = [0]*10
for i in range(10):
    make_set(p, i)

union_set(p, 2, 4)
union_set(p, 4, 6)
print(p)
```

```python
# DisJoint_Set by List2


class Disjoint_sets:
    def __init__(self, size=100):
        self.p = list(range(size))

    def find_set(self, x, flag=True):
        if self.p[x] == x:
            return x
        else:
            val = self.find_set(self.p[x])
            if flag:
                self.p[x] = val
            return val

    def union_set(self, x, y):
        self.p[self.find_set(y, flag=False)] = self.find_set(x)
```

```python
class DSNode:
    def __init__(self, val=None):
        self.val = val
        self.point = self
        self.rank = 1

class Disjoint_Sets:
    def __init__(self, size=100):
        self.p = [None]*size
        self.size = size
        for i in range(size):
            self.p[i] = DSNode(i)

    def find_set(self, x):
        tmp = self.p[x]
        if tmp.point is tmp:
            return tmp
        else:
            value = self.find_set(tmp.point.val)
            tmp.point = value
            return value

    def union_set(self, x, y):
        self.find_set(y).point = self.find_set(x)


    def __str__(self):
        a = [None]*self.size
        for i in range(self.size):
            a[i] = str(self.find_set(i).val)
        return " ".join(a)
```

```python
dj_set = Disjoint_Sets(7)
dj_set.union_set(2, 4)
dj_set.union_set(4, 5)
dj_set.union_set(3, 6)
dj_set.union_set(2, 3)
print(dj_set)
print(dj_set.find_set(0).val)
```
